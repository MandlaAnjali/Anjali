```python
x = np.arange(-9,10)
y =x**2
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("parabola")
plt.xticks(np.arange(-9,10,4))
plt.yticks(y)
```

```python
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("parabola")
plt.xlim(-5,5)
plt.ylim(10,50)
```

```python
x = np.arange(5,10)
plt.subplot(2,2,1)
plt.plot(x)
plt.title("X")
plt.subplot(2,2,2)
plt.plot(x,x+3,'o')
plt.title("X+3")
plt.subplot(2,2,3)
plt.plot(x,x*x)
plt.title("X*X")
plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title('logX')
```

```python
help(plt.subplot)
```

```python
x = np.arange(5,10)
plt.subplot(2,2,1,frameon=False)
plt.plot(x)
plt.title("X")
plt.subplot(2,2,2,projection='polar')
plt.plot(x,x+3,'o')
plt.title("X+3")
plt.subplot(2,2,3)
plt.plot(x,x*x)
plt.title("X*X")
plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title('logX')
```
